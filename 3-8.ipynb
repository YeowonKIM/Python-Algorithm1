<문제>

N*N 격자판으로 된 루피네 마당(N은 홀수). 격자단위에는 각각 말려야 하는 감의 개수가 있다.
해의 위치에 따라 잘 마르는 위치가 다르기 때문에 격자의 행을 왼쪽 또는 오른쪽으로 회전시키려 한다.
회전정보는 '회전할 행, 왼쪽과 오른쪽 중 어느 쪽으로 회전 시킬지(0은 왼쪽, 1은 오른쪽), 몇 만큼 회전시킬지'이다.
예를 들어, 회전정보가 2 0 3이면 2행을 왼쪽으로 3만큼 이동시킨다.

입력 : N, N*N 격자판에 들어갈 수(2차원 리스트), 회전 정보
출력 : 모래시계 모양으로 출력

```python
# 내 풀이
n = int(input())
arr =[list(map(int,input().split())) for _ in range(n)]
arr2 = []; arr3=[]
for i in range(n):
    for j in range(n):
        arr3.append(arr[i][j])
    arr2.append(arr3)
    arr3 = []
m = int(input())

for _ in range(m):
    a,b,c = map(int,input().split())
    if b==0:
        for i in range(n):
            if i-c<0:
                arr[a-1][i-c+n] = arr2[a-1][i]
            else:
                arr[a-1][i-c] =  arr2[a-1][i]
    if b==1:
        for i in range(n):
            if i+c<=n-1:
                arr[a-1][i+c] = arr2[a-1][i]
            else:
                arr[a-1][(i+c)-n] =  arr2[a-1][i]
tot = 0
s=0; e=n
for i in range(n):
    for j in range(s,e):
        # print(arr[i][j], end = " ")
        tot+=arr[i][j]
    if i<n//2:
        s+=1
        e-=1
    else:
        s-=1
        e+=1
print(tot)    
```

```python
# 수업풀이
n = int(input())
arr =[list(map(int,input().split())) for _ in range(n)]
m = int(input())
for i in range(m):
    a, b, c = map(int,input().split())
    if b == 0:
        for _ in range(c):
            arr[a-1].append(arr[a-1].pop(0))
    else:
        for _ in range(c):
            arr[a-1].insert(0,arr[a-1].pop())
#for x in arr:
    #print(x)
tot=0
s=0; e=n-1
for i in range(n):
    for j in range(s, e+1):
        tot+=arr[i][j]
    if i<n//2:
        s+=1
        e-=1
    else:
        s-=1
        e+=1
print(tot)
```
